```python
import re
```

```python
# X**2
```

```python
s = r'''2 в квадрате это 4'''
```

```python
regex_squared = re.compile(r'((?<=\s)|(?<=^)|(?<=\b))(?P<main>\d{1,})((?=\s)|(?=$)|(?=[,\.\?\!]))')
def num_squared(match):
    return str(int(match['main']) ** 2)

match = regex_squared.sub(num_squared, s)
print(match)
```

```python
# Замените все слова, которые начинаются на букву А в любом регистре, 
# на удалено(n), где n - длина удалённого слова.
```

```python
s2 = r'''А роза упала на лапу Азора.'''
```

```python
def counter_del(m):
    return f'удалено({len(m["main"])})'
regex = re.compile(r'((?<=\s)|(?<=^))(?P<main>[аА]\w*?)((?=\s)|(?=\b))')
match = regex.sub(counter_del, s2)
print(match)
```

```python
# Замена
# Нужно заменить все am на pm, а pm на am.
```

```python
s = r'''2:00 am 3:00 pm 10:00 pm 3:00 am 1:00 am  5:00 am 8:00 am 7:00 am 5:00 pm 8:00 pm 7:00 pm 6:00 pm 11:00 am 4:00 am 9:00 pm 6:00 am 12:00 am 12:00 pm 9:00 am 1:00 pm 11:00 pm 4:00 pm 2:00 pm 10:00 am'''
```

```python
def switcher_am(m):
    if(m['main'] == 'am'):
        return 'pm'
    else:
        return 'am'
regex = re.compile(r'((?<=\d)|(?<=\s)|(?<=^))(?P<main>am|pm)((?=\s)|(?=\b))')
match = regex.sub(switcher_am, s)
print(match)
```

```python
# MAC-адреса
# Скомпилируйте регулярное выражение, которое находит mac-адреса, и присвойте его переменной pattern. 
# Мак-адрес - специальный идентификатор, который присваивается сетевому адаптеру.
```

```python
s = r'''Мак-адрес моего друга:F0:98:9D:1C:93:F6. Мой мак-адрес: 0F:70:DE:55:60:19.'''
```

```python
regex = re.compile(r'(([0-9A-F]{2}:){5}[0-9A-F]{2})')
match = regex.findall(s)
print(match)
```
